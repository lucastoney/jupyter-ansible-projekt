# Funktionsnachweis JupyterLab

Dieses Notebook zeigt, dass JupyterLab erfolgreich mit Ansible bereitgestellt wurde.

```python
import getpass
import platform
import socket
from datetime import datetime

print("Hostname:", socket.gethostname())
print("Aktueller Benutzer:", getpass.getuser())
print("Python-Version:", platform.python_version())
print("Aktuelles Datum:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print("JupyterLab wurde erfolgreich mit Ansible bereitgestellt.")
```
